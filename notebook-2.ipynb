### Beeswarm 
ðŸ‘‰ Bonus

Grab your beeswarm code from the previous exercise

- modify the code to plot one plot per borough
- add a title and subtitle that helps tell a story with this chart

```python
%%R -w 1000 -h 200
library(ggbeeswarm)
ggplot(df) + 
    aes(x=pct_below_poverty, y=1) +
    geom_violin() +
    geom_beeswarm(cex = 1.5, alpha = 0.5)+
    facet_wrap(~borough)+
    labs(title = "Poverty rates across different boroughs",
         subtitle = "Each point represents the percentage of population below the poverty line in a borough",
         x = "Percentage below poverty line",
         y = "Borough") 
```
